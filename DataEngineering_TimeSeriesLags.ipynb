###Binary Encoding
 Transfrom Categorical Input Features to Binary Numbers

```python
# clear data (drop the years not all country don't have)
condition = df['Year'] < 1992
df=df.drop(df[condition].index)

# check the region codes
contrycodes= list(df['CountryCode'].unique())
regioncode= list(df['RegionCode'].unique())
print(contrycodes)
print(regioncode)
```

```python
# Helper function  a function to perform binary encoding
def binary_encode(column, new_name):
    categories = column.unique()
    max_val = len(categories) - 1  # Maximum value for encoding

    # Calculate the number of bits required to represent the maximum value
    num_bits = max_val.bit_length()
    print(num_bits)

    # Create a new DataFrame to hold the encoded columns
    encoded_df = pd.DataFrame()

    # Perform binary encoding
    for i in range(1, num_bits+1):
        col_name = new_name + str(i)
        encoded_df[col_name] = column.apply(lambda x: (categories.tolist().index(x) >> i) & 1)

    return encoded_df
```

```python
# Bindary Encoding of Region Codes
contrycodes= list(df['CountryCode'].unique())

binary_encoded1 = binary_encode(df['RegionCode'], 'RegionCode')

for column in binary_encoded1.columns:
    df[column] = binary_encoded1[column]
# Bindary Encoding  of Religion
binary_encoded2 = binary_encode(df['Religion'], 'Religion')
for column in binary_encoded2.columns:
    df[column] = binary_encoded2[column]
```

```python
# Bindary Encoding of ContryCodes
binary_encoded3 = binary_encode(df['CountryCode'], 'Country')
for column in binary_encoded3.columns:
    df[column] = binary_encoded3[column]
```

### Merge male, female, and unknwon sucide numbers for each year in each country

```python
#Combine male and female unknown rates
def sum_with_nan(values):
    if any(pd.isna(values)):
        return np.nan
    else:
        return np.sum(values)

# Define aggregation functions for each column
aggregation_functions = {
    'SuicideCount': sum_with_nan,
    'StdDeathRate': sum_with_nan
}

# Include all other columns by selecting the first value in each group
for column in df.columns:
  print(column)
  if column not in aggregation_functions and column != 'Year' and column != 'CountryCode':
        aggregation_functions[column] = 'first'

# Grouping the DataFrame by 'CountryCode' and 'Year' and aggregating using the defined functions
grouped_df = df.groupby(['CountryCode', 'Year']).agg(aggregation_functions).reset_index()
grouped_df['Sex'] = 'All'
all_df = grouped_df.copy()
#check count
all_df['SuicideCount'].value_counts(dropna=False)
```

```python
#Create Lagged DATA including
def create_lag(group, lag_n):
    group['lag{}'.format(lag_n)] = group['SuicideCount'].shift(lag_n)
    return group

grouped_data = all_df.groupby(['CountryCode'])
grouped_lag1_all_df = grouped_data.apply(create_lag,1)
lag1_all_df=grouped_lag1_all_df.reset_index(drop=True)
print(lag1_all_df)

grouped_lag1_all_df = lag1_all_df.groupby(['CountryCode'])
grouped_lag2_all_df=grouped_lag1_all_df.apply(create_lag,2)
lag2_all_df=grouped_lag2_all_df.reset_index(drop=True)

print(lag2_all_df)

grouped_lag2_all_df = lag2_all_df.groupby(['CountryCode'])
grouped_lag3_all_df=grouped_lag2_all_df.apply(create_lag,3)
lag3_all_df=grouped_lag3_all_df.reset_index(drop=True)

print(lag3_all_df)


grouped_lag3_all_df = lag3_all_df.groupby(['CountryCode'])
grouped_lag4_all_df=grouped_lag3_all_df.apply(create_lag,4)
lag4_all_df=grouped_lag4_all_df.reset_index(drop=True)

print(lag4_all_df)

grouped_lag4_all_df = lag4_all_df.groupby(['CountryCode'])
grouped_lag5_all_df=grouped_lag4_all_df.apply(create_lag,5)
lag5_all_df=grouped_lag5_all_df.reset_index(drop=True)

lag5_all_df['lag5'].isna()
```
